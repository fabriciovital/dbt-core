```python
import requests

# URL da API
url = "https://data.cityofnewyork.us/resource/erm2-nwe9.json"  # Exemplo de URL

response = requests.get(url)

if response.status_code == 200:
    # Parse da resposta JSON
    data = response.json()

    # Exibindo a estrutura do JSON
    print(data)  # Imprime a resposta JSON completa para inspeção

else:
    print(f"Erro na requisição: {response.status_code}")
```

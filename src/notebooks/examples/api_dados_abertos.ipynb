```python
import requests
import pandas as pd
from dotenv import load_dotenv
import os

# URL da API
load_dotenv()

API_DADOS_ABERTOS_URL=os.getenv('API_DADOS_ABERTOS_URL')
API_DADOS_ABERTOS_TOKEN=os.getenv('API_DADOS_ABERTOS_TOKEN')

# Cabeçalhos da requisição
headers = {
    'accept': 'application/json',
    'chave-api-dados-abertos': API_DADOS_ABERTOS_TOKEN
}

# Fazendo a requisição GET
response = requests.get(API_DADOS_ABERTOS_URL, headers=headers)

# Verificando se a requisição foi bem-sucedida
if response.status_code == 200:
    # Parse da resposta JSON
    data = response.json()

    # Criando uma lista para armazenar os dados
    dados = []

    # Extraindo informações desejadas e adicionando à lista
    for item in data:
        dados.append({
            'ID': item.get('id', 'ID não encontrado'),
            'Título': item.get('title', 'Título não encontrado'),
            'Nome': item.get('nome', 'Nome não encontrado'),
            'Data Atualização': item.get('dataAtualizacao', 'Data não encontrada'),
            'Nome da Organização': item.get('nomeOrganizacao', 'Organização não encontrada'),
        })

    # Criando um DataFrame a partir da lista
    df = pd.DataFrame(dados)

    # Exibindo o DataFrame
    display(df)  # use display em vez de print se estiver em um Jupyter Notebook

else:
    print(f'Erro na requisição: {response.status_code} - {response.text}')
```
